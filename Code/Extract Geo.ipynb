```python
import json
import pandas as pd
import zipfile
import shutil
import os

from google.colab import drive
drive.mount('/content/drive')
os.chdir("/content/drive/MyDrive/Data")
```

```python
def extract_geo(fname, countrycode):
    keylst = ['geo', 'place', 'user_location']
    for line in open(fname + '.json', 'r'):
        x = json.loads(line)
        for key in keylst:
            if x[f'{key}'] != {}:
                if x[f'{key}']['country_code'] == countrycode:
                    with open(fname + '.txt', 'a+', encoding='utf-8') as output:
                        output.write(str(x['tweet_id']+', '))
                        output.write(str(x[f'{key}']))
                        output.write(str('\n'))
                break
            else:
                continue
```

```python
Mondt = {"02":29, "03":31, "04":30, "05":1}
ctcd = 'gb'
path_txt = "/content/drive/MyDrive/Data/txt/"

for key, value in Mondt.items():
    for day_i in range(1, value+1):
        if day_i < 10:
            day_j = '0' + f'{day_i}'
        else:
            day_j = f'{day_i}'
        
        f_name = 'geo_2020-'+ key + '-' + day_j

        with zipfile.ZipFile('./' + f_name + '.zip', 'r') as zpf:
            zpf.extractall()
        
        extract_geo(f_name, ctcd)
        shutil.move(f_name + '.txt', path_txt+ f_name + '.txt')
        os.remove(f_name +'.json')
        os.remove(f_name +'.zip')
```
